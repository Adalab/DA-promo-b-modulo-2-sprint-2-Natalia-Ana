- Explorad la columna de country y reportad los resultados solo para aquellas categorías que incluyan el 70% de los datos.


```python
df.stb.freq(['country'], thresh=70, other_label="OTHER COUNTRIES")
```

CONCLUSIÓN: casi el 70% de los ataques se produjeron en USA, AUSTRALIA, SOUTH AFRICA Y PAPUA NEW GUINEA. Siendo USA el pais con mayor cantidad de casos seguido por Australia.

- Explorad la columna country en relación al sexo, filtrando de nuevo por aquellas categorías que incluyan el 70% de los datos.


```python
df.stb.freq(['sex', 'country'], thresh=70, other_label="OTHER")
```

CONCLUSIÓN: La mayoría de los ataques se han producido en hombres en todos los países y el pais con mayor número de ataques en mujeres es USA. Esto es esperable ya que USA es el pais con mayor cantidad de ataques totales a nivel mundial. 

- Haced los mismo para la columna activity en relación a la columna fatal. En este caso filtrad y mostrad los datos que incluyan el 60% de los datos.


```python
df.stb.freq(['activity', 'fatal'], thresh=70, other_label="Other")
```
